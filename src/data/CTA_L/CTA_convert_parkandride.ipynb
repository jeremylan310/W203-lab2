```python
import csv
```

```python
# First, split into { station_name: data_rows }
partitioned = {}
station_name = ''

with open('CTA_ParkandRide_raw.csv') as csvfile:
    csvreader = csv.reader(csvfile)

    for row in csvreader:
        if len(row) > 1 and row[0].startswith('POLYGON'):
            
            # Add all datarows
            if station_name:
                partitioned[station_name] = data_rows

            station_name = row[1]
            data_rows = []
        else:
            data_rows.append(row[0].strip())

partitioned
```

```python
# Then, reorg the data_rows into dictionaries:
cleaned = {}
for key, value in partitioned.items():
    contents = {}
    for row in value:
        items = row.split('   ')
        if len(items) > 1:
            field, field_value, = items
            if field != 'Field Name':
                contents[field] = field_value

    cleaned[key] = contents
cleaned
```

```python
# Get all column names
columns = {name for key, value in cleaned.items() for name, field_value in value.items()}
```

```python
with open('CTA_ParkandRide_cleaned.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(columns)
    for station, fields in cleaned.items():
        csvwriter.writerow([fields[c] for c in columns])
```
